```python
#cargargamos el archivo csv en un dataframe
import pandas as pd
orders_df = pd.read_csv('../Data/orders.csv')
```

```python
#Mostramos las primeras filas del dataframe para analizar su contenido
orders_df.head(20)
```

```python
#Al analizar el df, observamos que hay una columna llammada order date, la cual contiene
# la informacion de la fecha en la que se realizo el pedido, pero esta columna no se encuentra en el formato de
#fecha, por lo que se procede a cambiar el formato utilizando la funcion to_datetime.
#Ademas de observar que la columna ship mode, contiene la informaci√≥n del modo de envio, 
# #pero contiene valores con referena nulos
orders_df = pd.read_csv('../Data/orders.cfsv', na_values=['Not Available','unknown'])
orders_df['Ship Mode'].unique()
```

```python
#sustituimos las columnas por nombres mas cortos y sin espacios, para facilitar su manejo
orders_df.columns = orders_df.columns.str.lower().str.replace(' ','_')
orders_df.head()
```

```python
#creamos una nueva columna para almacenar el descuento que se aplica al precio 
#creamos una nueva columna para almacenar el precio de venta, que se obtiene restando el descuento al precio de lista  
#creamos una nueva columna para almacenar el beneficio, que se obtiene restando el precio de venta al precio de lista 
orders_df['discount'] = (orders_df['list_price']* orders_df['discount_percent']) / 100 
orders_df['sales_price'] = orders_df['list_price'] - orders_df['discount']
orders_df['profit'] = orders_df['sales_price'] - orders_df['cost_price']
orders_df.head()
```

```python
#modificamos el formato de la columna order_date a formato fecha utilizando la funcion to_datetime
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'],format='%Y-%m-%d')
orders_df.dtypes
orders_df.head()
```

```python
#eliminamos las columnas que no son necesarias para el analisis, como 
orders_df.drop(columns=['cost_price','list_price','discount_percent'], inplace=True)
orders_df
```
